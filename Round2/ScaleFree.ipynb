# *Importing required libraries*

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

# *Creating Scale Free Graph using Barabasi  Albert model*

```python

G= nx.barabasi_albert_graph(250,4)
pos=nx.spring_layout(G)
nx.draw(G, with_labels=True)
plt.show()
```

# *Assigning random weights to edges*

```python

edges=list(G.edges)
for u,v in edges:
    G.add_weighted_edges_from([(u, v, random.randint(1,10))])
    #print(u,v,":",G[u][v]['weight'])
```

# *Normalizing the weights to act as  probability*

```python
edges=list(G.edges)
d={}
for u,v in edges:
    if u not in d:
        d[u]=G[u][v]['weight']
    else:
        d[u]+=G[u][v]['weight']

for u,v in edges:
    G[u][v]['weight']=G[u][v]['weight']/d[u]
    print(u,v,":",G[u][v]['weight'])
```

# *Applying The ICM algorithm*

```python
import numpy as np
def IC(Networkx_Graph,times):
    mx=0
    for i in range(times):
        ans=[]
        curr=[]
        if(i not in ans):
            ans.append(i)
        targets = [n for n in G.neighbors(i)]
        for k in targets:
            if(np.random.random()<G[i][k]['weight'] and k not in ans):
                ans.append(k)
                curr.append(k)                
        steps=0
        for m  in curr:
            targets = [n for n in G.neighbors(m)]
            for k in targets:
                if(k not in ans and k not in curr and np.random.random()<G[m][k]['weight'] ):
                    ans.append(k)
                    curr.append(k)
            steps=steps+1
        
        
        mx=max(len(ans),mx)      
        print("Inital activated node :",i,end='\n')
        print("Activated nodes:",ans,end='\n')
        print("Count of activated nodes:",len(ans),end='\n')
        print("Steps taken:",steps,end='\n')
        print('\n')
    return mx 


print("Maxinum No of activated nodes  is:",IC(G,5))
```
